```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("cgpa_package_200_students.csv")
```

```python
plt.scatter(df['CGPA'],df['Package_LPA'])
plt.title("CGPA vs Package_LPA")
plt.xlabel("CGPA")
plt.ylabel("Package_LPA")
plt.grid(True)
plt.show()
```

```python
X = df[["CGPA"]] # there is a use of the double [[]] because it will denote the use of the X cordinate ( input )
Y = df["Package_LPA"]
X,Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_test , y_test
```

```python
model.predict(X_test)
```

```python
plt.scatter(df['CGPA'],df['Package_LPA'])
plt.title("CGPA vs Package_LPA", color = "Red")
plt.plot(X_test, model.predict(X_test), color = "Black" ) #accuracy based upon the model trained
plt.xlabel("CGPA")
plt.ylabel("Package_LPA")
plt.grid(True)
plt.show()
```
